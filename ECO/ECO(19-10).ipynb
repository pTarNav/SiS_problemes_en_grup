```python
import numpy as np
import scipy.io.wavfile
import matplotlib.pyplot as plt
from scipy.signal import *
from math import sqrt
```

```python
plt.plot(scipy.io.wavfile.read('eco1.wav')[1])
```

```python
frame_rate = scipy.io.wavfile.read('eco1.wav')[0]
y = scipy.io.wavfile.read('eco1.wav')[1].astype(np.intp)
```

```python
Ryy = convolve(y, y[::-1])
Ryy = list(Ryy)
peaks = []
for i in range(1, len(Ryy)-1):
    prev = Ryy[i-1]
    curr = Ryy[i]
    nxt = Ryy[i+1]
    if prev < curr and curr > nxt:
        peaks.append([i, curr])
peaks.sort(reverse=True, key=lambda x : x[1] )
Ryy0 = peaks[0][1].item()
pos_Ryy0 = peaks[0][0]
RyyN = peaks[1][1].item()
pos_RyyN = peaks[1][0]
N = abs(pos_Ryy0 - pos_RyyN)

plt.plot(Ryy)
N, Ryy0, RyyN
```

```python
a1 = (Ryy0 + sqrt(Ryy0**2 - 4*RyyN**2))/(2*RyyN)
a2 = (Ryy0 - sqrt(Ryy0**2 - 4*RyyN**2))/(2*RyyN)
a1, a2
```

```python
b = a2
h2 = np.zeros(len(y))
for i in range(len(y)):
    if i%N == 0:
        h2[i] = (-b)**(i//N)
        
# OJO! Audio s'escolta MOLT FORT!    

z = convolve(y,h2, mode='full')
scipy.io.wavfile.write('sin_eco1_.wav',frame_rate, z)
plt.plot(z), b
```
